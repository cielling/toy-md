```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

```python
# eps = 1.0
# sigma = 1.0
eps = 30682.0
sigma = 2.928
r_min = m.pow(2.0, 1/6.) * sigma
print(r_min)
```

```python
2.5*r_min
```

```python
r = np.arange(1,500*sigma)/100
lj = 4*eps*( np.power(sigma/r, 12.0) - np.power(sigma/r, 6.0))
```

```python
plt.plot(r, lj)
plt.ylim(-31000.1, 10000.0)
plt.show()
```
